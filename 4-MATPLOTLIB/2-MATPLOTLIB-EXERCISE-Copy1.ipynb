```python
import numpy as np
x=np.arange(0,100)
y=x*2
z=x**2
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.show
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.set_title('Title')
ax.set_xlabel('X label')
ax.set_ylabel('Y label')
ax.plot(x,y)
```

```python
fig=plt.figure()
ax1=fig.add_axes([0,0,1,1])
ax2=fig.add_axes([0.2,0.5,0.2,.2])
ax1.plot(x,y)
ax2.plot(y,x)
```

```python
fig=plt.figure()
ax1=fig.add_axes([0,0,1,1])
ax2=fig.add_axes([0.2,0.5,.4,.4])
ax1.plot(x,z)
ax2.plot(x,y)
ax2.set_xlim(20,22)
ax2.set_ylim(30,50)
```

```python
plt.plot(x,y,z,'r')
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2,figsize=(4,5))
axes[0].plot(x,y,lw=3,ls=':')

axes[1].plot(x,z,'r',lw=4,ls='--')
plt.tight_layout()
```
