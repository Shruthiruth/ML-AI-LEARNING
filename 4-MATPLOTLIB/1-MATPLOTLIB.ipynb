```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(1,6)
y = x**2

plt.subplot(1,2,1)
plt.plot(x,y)
plt.title("First")

plt.subplot(1,2,2)
plt.plot(y,x)
plt.title("Second")

plt.show()
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2)

# for current_ax in axes:
#     current_ax.plot(x,y)


axes[0].plot(x,y)
axes[1].plot(y,x)
axes[0].set_title('first plot')
axes[1].set_title('second plot')
plt.tight_layout() # overlapping issue
```

```python
# figure size and DPI--dot per inch
```

```python
fig=plt.figure(figsize=(8,2)) #width,height

ax=fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig,axes=plt.subplots(nrows=2,ncols=1,figsize=(8,2)) #width,height

axes[0].plot(x,y)
axes[1].plot(y,x)

plt.tight_layout()
```

```python
fig=plt.figure()

axes=fig.add_axes([0.1,0.1,0.8,0.8]) #left,bottom,width,height

# axes.plot(x,y)
# axes.plot(y,x)
axes.plot(x,x**2,label='X squared')
axes.plot(x,x**3,label='X cubed')

# axes.legend(loc=(0.1,0.1))

axes.legend(loc=0)
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y,color='#FF8C00')
ax.plot(x,y,color='purple',lw=20,alpha=0.5,linestyle='--',marker='*',markersize=20) #linewidth-lw
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y,color='#FF8C00')
ax.plot(x,y,color='purple',lw=2,ls=':',marker='o',markersize=20,markerfacecolor='yellow',markeredgewidth=3,markeredgecolor='green') #linewidth-lw ,ls--linestyle
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y,color='purple',lw=2,ls='--') #linewidth-lw ,ls--linestyle
ax.set_xlim([0,1])
ax.set_ylim([0,2])
```
