```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import plotly.graph_objects as go
import plotly.express as px
from plotly.offline import init_notebook_mode,download_plotlyjs,plot,iplot
init_notebook_mode(connected=True)
```

```python
df=pd.read_csv('2014_World_Power_Consumption')
```

```python
df.head()
```

```python
data=go.Choropleth(
    locations=df['Country'],
    locationmode='USA-states',
    z=df['Power Consumption KWH'],
    text=df['Text'],
    colorscale='YlOrRd',
     marker=dict(
        line=dict(color='rgb(255,255,255)', width=2)
    ),
    colorbar_title='2014_World_Power_Consumption'
    
)

layout = go.Layout(
    title='2014_World_Power_Consumption',
    geo=dict(
        projection=dict(type='natural earth'),
        showframe=False
    )
)

fig1 = go.Figure(data=[data], layout=layout)
fig1.show()
```

```python
df1=pd.read_csv('2012_Election_Data')
                
```

```python
df1.head()
```

```python
data = go.Choropleth(
    locations=df1['State Abv'],
    locationmode='USA-states',
    z=df1['Voting-Age Population (VAP)'],
    text=df1['State'],
    colorscale='Viridis',
    marker=dict(
        line=dict(color='rgb(255,255,255)', width=2)
    ),
    colorbar_title='2012 Election Data'
)

layout = go.Layout(
    title='2012 Election Data',
    geo=dict(
        scope='usa',
        projection=dict(type='natural earth'),
        showlakes=True,
        lakecolor='rgb(85,173,240)'
    )
)

fig2 = go.Figure(data=[data], layout=layout)
fig2.show()
```
