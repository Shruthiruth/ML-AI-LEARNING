```python
import plotly.graph_objects as go

data = go.Choropleth(
    colorscale='YlOrRd',   # Correct colorscale name
    locations=df['code'],
    locationmode='USA-states',
    z=df['total exports'],
    text=df['text'],
    marker=dict(
        line=dict(color='rgb(255,255,255)', width=2)
    ),
    colorbar_title='Millions USD'
)

layout = go.Layout(
    title='2011 US Agriculture Exports by State',
    geo=dict(
        scope='usa',
        showlakes=True,
        lakecolor='rgb(85,173,240)'
    )
)

fig1 = go.Figure(data=[data], layout=layout)
fig1.show()
```

```python
df=pd.read_csv('2014_world_GDP')
```

```python
df.head()
```

```python
data = go.Choropleth(
    locations=df['CODE'],
    z=df['GDP (BILLIONS)'],
    text=df['COUNTRY'],
    locationmode='ISO-3',
    colorscale='YlOrRd',
    marker=dict(
        line=dict(color='white', width=0.5)
    ),
    colorbar_title='GDP in Billions USD'
)

layout = go.Layout(
    title='2014 Global GDP',
    geo=dict(
        projection=dict(type='natural earth'),
        showframe=False
    )
)

fig = go.Figure(data=[data], layout=layout)
fig.show()
```
