```python
# Create a function that grabs the email website domain from a string in the form: **

# user@domain.com

# So for example, passing "user@domain.com" would return: domain.com

def domain(x):
    output = x.split("@")
    return output[1]

domain("user@domain.com")
```

```python
# Create a basic function that returns True if the word 'dog' is contained in the input string. Don't worry about edge cases like a punctuation being attached to the word dof,but do account for captialization

# findDog('Is there a dog here?")

def dogfunc(x):
    ch=x.split()
    if "dog" in ch:
        return True
    return False

dogfunc("Is there a dog here?")
    
```

```python
# * Create a function that counts the number of times the word "dog" occurs in a string. Again ignore edge cases,*


def countDog(words):
    count=0
    ch=words.split()
    
    for word in ch:
        if word == "dog" :
            count += 1


    return count
            
print(countDog("This dog runs faster than the other dog dudel"))
```
